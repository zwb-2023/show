### 生成预测效果

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generate sample data (replace with your actual data)
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Features
y = 2.5 * X.squeeze() + np.random.normal(0, 2, 100)  # Actual values

# Train model and predict
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Create visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y, y=y_pred, alpha=0.6, label='Data Points')
plt.plot([min(y), max(y)], [min(y), max(y)], 'r--', lw=2, label='Ideal Prediction')

# Add regression line
sns.regplot(x=y, y=y_pred, scatter=False, 
            line_kws={'color':'green', 'lw':2}, 
            label='Regression Fit')

plt.title('Prediction vs Actual Performance', fontsize=14)
plt.xlabel('Actual Values', fontsize=12)
plt.ylabel('Predicted Values', fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

# Add metrics
r2 = r2_score(y, y_pred)
plt.annotate(f'R² = {r2:.3f}', xy=(0.05, 0.9), xycoords='axes fraction',
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

plt.tight_layout()
plt.show()
```

### 生成预测伪色图

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图片
image = cv2.imread('M-3-1-1.bmp', cv2.IMREAD_GRAYSCALE)

# 生成伪色图
colormap_image = cv2.applyColorMap(image, cv2.COLORMAP_DEEPGREEN)

# 添加文字
font = cv2.FONT_HERSHEY_SIMPLEX
text = "一号位置竹子含水量低，二号位置含水量高，三号位置含水量中等"
cv2.putText(colormap_image, text, (10, 50), font, 1, (255, 255, 255), 2, cv2.LINE_AA)

# 显示图片
plt.imshow(cv2.cvtColor(colormap_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# 保存生成的图像
cv2.imwrite('output_image.jpg', colormap_image)
```
